```python
from regression import *
import numpy as np
import matplotlib.pyplot as plt
import time

%load_ext autoreload
%autoreload 2
```

```python
original_data_file = "housing.txt"
processed_data_file = "normalized.txt"
```

```python
normalize_file(original_data_file,processed_data_file)
```

```python
x,y,mean,std = load_data(processed_data_file)
```

```python
max_limit = range(10,90,10)
alphas = [0.01,0.1,0.3]
for alpha in alphas:
    losses = []
    for limit in max_limit:
        losses.append(train(x,y,alpha,limit)[0])
    plt.plot(max_limit,losses)
    plt.xlabel('Number of iteration')
    plt.ylabel('J(w)')
    plt.title('J(w) as a function of the number of iteration (alpha='+str(alpha)+")")
    plt.show()
```

```python
alphas = [0.05,0.5,1.5]
for alpha in alphas:
    print("Alpha=",alpha)
    loss = train(x,y,alpha,limit,True)[0]
    print(loss)
```
